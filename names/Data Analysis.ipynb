Let's plot the top 5 names for male and females during 2017

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.75, 0.75])
ax.plot(male.Name[0:5], male.Count[0:5],'go-', linestyle='dashed', label='2017')
ax.set_xlabel('Name')
ax.set_ylabel('Number of name occurrences')
ax.set_title('Top male baby names for year 2017')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

```python
starting_year = 1990
ending_year = 2017
years = []
#Create a list for the years we want to obtain top baby names
for i in range((ending_year-starting_year)+1):
    years.append(starting_year+i)
```

```python
fig_M = plt.figure()
ax = fig_M.add_axes([0.1, 0.1, 2.25, 0.75])

fig_F = plt.figure()
ax2 = fig_F.add_axes([0.1, 0.1, 2.25, 0.75])

#TODO: Get count of top names ??? 
for year in years:
    current_year = 'yob'+str(year)+'.csv'
    current_output = pd.read_csv(current_year)
    #get male and female names
    male = current_output[current_output.Sex == 'M']
    female = current_output[current_output.Sex == 'F']
    
    ax.plot(male.Name[0:5], male.Count[0:5],'o', label=str(year))
    ax.set_xlabel('Name')
    ax.set_ylabel('Number of name occurrences')
    ax.set_title('Top male baby names')
    
    ax2.plot(female.Name[0:5], female.Count[0:5],'o', label=str(year))
    ax2.set_xlabel('Name')
    ax2.set_ylabel('Number of name occurrences')
    ax2.set_title('Top female baby names')
    
    ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    ax2.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.show()
    
```

Now, let's find the number of babies born each year

```python
yearly_population = dict()
for year in years:
    current_year = 'yob'+str(year)+'.csv'
    current_output = pd.read_csv(current_year)
    baby_total = 0
    for current_name_count in current_output.Count:
        baby_total += current_name_count
    yearly_population[year] = baby_total
yearly_population
```

```python
plt.bar(yearly_population.keys(),yearly_population.values())
plt.xlabel('Year')
plt.ylabel('Number of births')
plt.axis()
plt.title('Yeary births in the U.S. from 1990 to 2017')
plt.show()
```
